__Q5: Which year has the highest percentage of blowout games, which is defined as a winning margin of at least 30 points?__

```python
df.head()
```

```python
df.groupby(['Season']).count()
```

```python
blowout = df[df.score_diff>=30]
```

```python
blowout.head()
```

```python
blowout.groupby(['Season']).count()
```

```python
blowout.groupby(['Season']).count() / df.groupby(['Season']).count()
```

```python
percentageof = blowout.groupby(['Season']).count() / df.groupby(['Season']).count()
```
